```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
resumes = pd.read_csv("../data/UpdatedResumeDataSet.csv")
jobs = pd.read_csv("../data/job_roles.csv")
```

```python
print("Resume Dataset Shape:", resumes.shape)
print("Job Dataset Shape:", jobs.shape)
```

```python
resumes.head()
```

```python
jobs.head()
```

```python
print("Resume Columns:", resumes.columns)
print("Job Columns:", jobs.columns)
```

```python
print("Resume Missing Values:\n", resumes.isnull().sum())
print("Job Missing Values:\n", jobs.isnull().sum())
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y=resumes['Category'])
plt.title("Resume Category Distribution")
plt.show()
```

```python
resumes['text_length'] = resumes['Resume'].apply(len)

plt.figure(figsize=(8,5))
sns.histplot(resumes['text_length'], bins=50)
plt.title("Resume Text Length Distribution")
plt.show()
```

```python
jobs['jd_length'] = jobs['Required Skills'].astype(str).apply(len)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(jobs['jd_length'], bins=50)
plt.title("Job Skills Length Distribution")
plt.show()
```
